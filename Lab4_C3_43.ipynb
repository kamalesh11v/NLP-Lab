To study and perform histogram processing in image enhancement
1. plot a histogram of grayscale and color image
2. Contrast and brightness stretching of of grayscale image.
3. Perform Histogram equalization on grayscale image

```python
# 1) plot a histogram of grayscale and color image

import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
gray_img = cv2.imread('/content/cameraman.jpg', cv2.IMREAD_GRAYSCALE)
print("greyscale image")
cv2_imshow(gray_img)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(gray_img.ravel(), 256, [0, 256])
plt.title('Grayscale Histogram')
color = ('b', 'g', 'r')

```

```python
# 1) plot a histogram of grayscale and color image

import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
color_img = cv2.imread('/content/lenacolor.png')
print("color image")
cv2_imshow(color_img)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 2)
for i, col in enumerate(color):
    histr = cv2.calcHist([color_img], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
plt.title('Color Histogram')
plt.show()
```

```python

# 2. Contrast and brightness stretching of of grayscale image.
# Contrast stretching
# multiplying a constant
constant = 1.5  # Adjust this value for different contrast levels
contrast_stretched = cv2.multiply(gray_img, constant)
contrast_stretched = np.clip(contrast_stretched, 0, 255).astype(np.uint8)  # Ensure pixel values are within [0, 255]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(contrast_stretched, cmap='gray')
plt.title('Contrast Stretched Image')


plt.subplot(1, 2, 2)
plt.hist(contrast_stretched.ravel(), 256, [0, 256])
plt.title('Contrast Stretched Histogram')

plt.show()

cv2_imshow(contrast_stretched)
```

```python

# 3. Perform Histogram equalization on grayscale image
equalized_img = cv2.equalizeHist(gray_img)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(equalized_img, cmap='gray')
plt.title('Histogram Equalized Image')

plt.subplot(1, 2, 2)
plt.hist(equalized_img.ravel(), 256, [0, 256])
plt.title('Histogram Equalized Histogram')

plt.show()

cv2_imshow(equalized_img)
```
