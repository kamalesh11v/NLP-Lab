#### To study and perform image morphology using following techniques :    
#### i) Dilation
#### ii) Erosion
#### iii) Opening
#### iv) Closing

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/drive/MyDrive/DIVP/Letter H.jpg", cv2.IMREAD_GRAYSCALE)
kernel = np.ones((3, 3), np.uint8)

dilation1 = cv2.dilate(image, kernel, iterations=20)
dilation2 = cv2.dilate(image, kernel, iterations=30)

height, width = image.shape
depth = 1

print("Original Image - Dimensions: Height = {}, Width = {}, Depth = {}".format(height, width, depth))
cv2_imshow(image)

height1, width1 = dilation1.shape
height2, width2 = dilation2.shape

print("\nDilated Image with 10 iterations - Dimensions: Height = {}, Width = {}, Depth = {}".format(height1, width1, depth))
cv2_imshow(dilation1)

print("\nDilated Image with 20 iterations - Dimensions: Height = {}, Width = {}, Depth = {}".format(height2, width2, depth))
cv2_imshow(dilation2)
```

```python
# Erosion

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/drive/MyDrive/DIVP/Letter H.jpg", cv2.IMREAD_GRAYSCALE)
kernel = np.ones((3, 3), np.uint8)

erosion1 = cv2.erode(image, kernel, iterations=10)
erosion2 = cv2.erode(image, kernel, iterations=20)

height, width = image.shape
depth = 1

print("Original Image - Dimensions: Height = {}, Width = {}, Depth = {}".format(height, width, depth))
cv2_imshow(image)

height1, width1 = erosion1.shape
height2, width2 = erosion2.shape

print("\nEroded Image with 10 iterations - Dimensions: Height = {}, Width = {}, Depth = {}".format(height1, width1, depth))
cv2_imshow(erosion1)

print("\nEroded Image with 20 iterations - Dimensions: Height = {}, Width = {}, Depth = {}".format(height2, width2, depth))
cv2_imshow(erosion2)
```

```python
#Opening

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/drive/MyDrive/DIVP/Letter H.jpg")
print("Original Image \n")
cv2_imshow(image)
w,h,d = image.shape
print(w,h,d)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel, iterations=40)
print("\n Opening Image with 40 iterations")
cv2_imshow(opening)
```

```python
#Closing

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/drive/MyDrive/DIVP/Letter H.jpg")
print("Original Image \n")
cv2_imshow(image)
w,h,d = image.shape
print(w,h,d)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel, iterations=35)
print("\n Closing Image with 35 iterations")
cv2_imshow(closing)
```
